```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

from src_regi.config import artifacts_dir, load_config
from src_regi.data import prepare_data
from src_regi.datasets import build_dataloaders
from src_regi.latent_cycle_gan import LatentCycleGAN
from src_regi.timegan import TimeGAN
from src_regi.utils import get_device, load_checkpoint
from src_regi.world_encoder import WorldEncoder

cfg = load_config(PROJECT_ROOT / "config_regi.json")
device = get_device()
ckpt = load_checkpoint(artifacts_dir(cfg) / "phase1_best.pt", map_location=device)
prepared = prepare_data(cfg)
seq_len = cfg["data"]["sequence_length"]
num_features = len(prepared.feature_columns)
mcfg = cfg["model"]

encoder = WorldEncoder(num_features, seq_len, mcfg["d_model"], mcfg["d_latent"], mcfg["nhead"], mcfg["num_layers"], mcfg["dropout"]).to(device)
encoder.load_state_dict(ckpt["world_encoder"])
encoder.eval()

loader = build_dataloaders(prepared, seq_len, 64, phase="phase1")["val"]
zs = []
regimes = []
years = []

with torch.no_grad():
    for batch in loader:
        zs.append(encoder(batch["x"].to(device)).cpu().numpy())
        # Assume 'regime' and 'year' are keys in batch if available; add fallback if not
        if "regime" in batch:
            batch_regime = batch["regime"].cpu().numpy()
            # If batch_regime is 1D (samples,): align with zs[-1] shape 0th axis
            if batch_regime.shape[0] == zs[-1].shape[0]:
                regimes.append(batch_regime)
            else:
                regimes.append(np.repeat(batch_regime, zs[-1].shape[0] // len(batch_regime)))
        else:
            regimes.append(np.full((zs[-1].shape[0],), np.nan))
        if "year" in batch:
            batch_year = batch["year"].cpu().numpy()
            if batch_year.shape[0] == zs[-1].shape[0]:
                years.append(batch_year)
            else:
                years.append(np.repeat(batch_year, zs[-1].shape[0] // len(batch_year)))
        else:
            years.append(np.full((zs[-1].shape[0],), np.nan))

Z = np.vstack(zs)
regimes = np.concatenate(regimes)
years = np.concatenate(years)

Z2 = PCA(n_components=2).fit_transform(Z)

plt.figure(figsize=(6, 5))

# Use regime for coloring if nontrivial, otherwise fallback to year
if not np.all(np.isnan(regimes)):
    scatter = plt.scatter(Z2[:, 0], Z2[:, 1], s=8, alpha=0.5, c=regimes, cmap='tab10', label="Regime")
    plt.colorbar(scatter, label="Regime")
    plt.title("Validation macro latents (PCA, colored by regime)")
elif not np.all(np.isnan(years)):
    scatter = plt.scatter(Z2[:, 0], Z2[:, 1], s=8, alpha=0.5, c=years, cmap='viridis', label="Year")
    plt.colorbar(scatter, label="Year")
    plt.title("Validation macro latents (PCA, colored by year)")
else:
    plt.scatter(Z2[:, 0], Z2[:, 1], s=8, alpha=0.5)
    plt.title("Validation macro latents (PCA)")

plt.tight_layout()
plt.show()
```

# In the PCA scatter plots above, the color of each point represents additional categorical or temporal information about each sample:
# - In the "Validation macro latents (PCA, colored by regime)" plot, different colors correspond to different regime IDs for each sample (e.g., different market regimes, environments, or clusters). The colors (from the 'tab10' colormap) let you visually distinguish which regime each point belongs to.
# - If regime information is not available, the code instead colors points by 'year' (using the 'viridis' colormap), so different years appear as a gradient from one color to another. This helps you see if year is related to the spatial structure in the latent space.
# - If neither regime nor year info is present, the plot appears in a single color (no coloring).
# - For the training and test scatter plots, all points are a single color ('tab:blue' for train, 'tab:green' for test); here, color simply distinguishes the dataset, not any semantic property.

```python
# Plot PCA projection for train latents
loader_train = build_dataloaders(prepared, seq_len, 64, phase="phase1")["train"]
zs_train = []
with torch.no_grad():
    for batch in loader_train:
        zs_train.append(encoder(batch["x"].to(device)).cpu().numpy())
Z_train = np.vstack(zs_train)
Z2_train = PCA(n_components=2).fit_transform(Z_train)
plt.figure(figsize=(6, 5))
plt.scatter(Z2_train[:, 0], Z2_train[:, 1], s=8, alpha=0.5, color='tab:blue')
plt.title("Train macro latents (PCA)")
plt.tight_layout()
plt.show()

# If you have a test split, plot for test latents
if "test" in build_dataloaders(prepared, seq_len, 64, phase="phase1"):
    loader_test = build_dataloaders(prepared, seq_len, 64, phase="phase1")["test"]
    zs_test = []
    with torch.no_grad():
        for batch in loader_test:
            zs_test.append(encoder(batch["x"].to(device)).cpu().numpy())
    Z_test = np.vstack(zs_test)
    Z2_test = PCA(n_components=2).fit_transform(Z_test)
    plt.figure(figsize=(6, 5))
    plt.scatter(Z2_test[:, 0], Z2_test[:, 1], s=8, alpha=0.5, color='tab:green')
    plt.title("Test macro latents (PCA)")
    plt.tight_layout()
    plt.show()
```
